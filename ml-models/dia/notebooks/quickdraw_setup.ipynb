# Feature - DIA || QuickDraw Data

## Import Relevant Libraries

```python
import os
import urllib.request
```

## Define Paths

```python
BASE = "https://storage.googleapis.com/quickdraw_dataset/full/numpy_bitmap/"
OUT_DIR = "quickdraw_npy"
os.makedirs(OUT_DIR, exist_ok=True)
```

## Define Categories to Extract

```python
categories = [
    "arm","beard","brain","ear","eye","eyeglasses","face","finger","foot",
    "goatee","helmet","knee","leg","mermaid","moustache","mouth","nose",
    "shoe","skull","smiley face","t-shirt","underwear"
]
```

## Extract Data

```python
def qd_filename(cat):
    # quickdraw file names use spaces 그대로 (URL-encoded by urllib)
    return cat + ".npy"
```

```python
for cat in categories:
    url = BASE + qd_filename(cat)
    path = os.path.join(OUT_DIR, qd_filename(cat))
    if not os.path.exists(path):
        print("Downloading:", cat)
        urllib.request.urlretrieve(url, path)
```

```python
from urllib.parse import quote

for cat in ["smiley face","t-shirt","underwear"]:
    local_path = os.path.join(OUT_DIR, f"{cat}.npy")

    # skip if already downloaded
    if os.path.exists(local_path):
        print(f"Skipping (already exists): {cat}")
        continue

    url = BASE + quote(f"{cat}.npy")

    print(f"Downloading: {cat}")
    urllib.request.urlretrieve(url, local_path)
```
